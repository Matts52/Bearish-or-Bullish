```python

sent_model_190 = pd.read_csv('var_190_sent_scale_V5_DS2_1.csv')
sent_model_190 = sent_model_190.drop(['Unnamed: 0', 'VIXCLS'], axis=1)
df_polarity = sent_model_190.mean()*100
df_polarity = df_polarity.to_frame(name="mean_sentiment")

df_polarity = df_polarity.sort_values("mean_sentiment")
df_polarity['topic'] = df_polarity.index
print("top pos topics:")
for i in df_polarity['topic'].iloc[-10:]:
    print(i)
print("top neg topics:")
for i in df_polarity['topic'].iloc[:10]:
    print(i)
sns.set(style="white", rc = {'figure.figsize':(300,150)}, font_scale = 4.5)

ax = sns.barplot(x="topic", y= "mean_sentiment", data=df_polarity, orient='v', 
                 order=df_polarity.sort_values('mean_sentiment').topic)
ax.set_xticklabels(ax.get_xticks(), size = 50)
ax.set_yticklabels(ax.get_yticks(), size = 150)
ax.set_ylabel('Mean Polarity', fontsize = 200)
ax.axes.xaxis.set_ticks([]) # remove xticks
ax.set_xlabel('Topic', fontsize = 200)
ax.set_title('Sentiment Weighted Topic Attention Polarity', fontsize = 250)
# red_patch = mpatches.Patch(color='red', label='The red data')
# plt.legend(handles=[red_patch, ])
legend_pos = plt.legend(title='Top Positive Topics', fontsize=200, bbox_to_anchor=(0.3, 1), frameon=True, labels=[topics190[46], topics190[135], topics190[102], topics190[81], topics190[22], topics190[86], topics190[16], topics190[88], topics190[174], topics190[94]],   prop={'size': 150}) # color=['red', 'green'],
#legend=plt.legend(list,loc=(1.05,0.05), title=r'$\bf{Title}$') #Legend: list, location, Title (in bold)

# leg = Legend(tit, lines[2:], ['line C', 'line D'],
#              loc='lower right', frameon=False)
# ax.add_artist(leg);
legend_neg = plt.legend(title='Top Negative Topics', fontsize=200, bbox_to_anchor=(0.15, 1), frameon=True, labels=[topics190[127], topics190[101], topics190[85], topics190[114], topics190[189], topics190[51], topics190[136], topics190[44], topics190[140], topics190[35]],  prop={'size': 150})
#ax.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
#pyplot.legend([l[0] for l in plot_lines], parameters, loc=4)
legend_pos.get_title().set_fontsize('200') #legend 'Title' fontsize
legend_neg.get_title().set_fontsize('200')

plt.gca().add_artist(legend_neg)
plt.gca().add_artist(legend_pos)

#plt.savefig('Sentiment_Weighted_Topic_Attention_Polarity.png', bbox_inches='tight')
plt.show()

figure = ax.get_figure()    
figure.savefig('Sentiment_Weighted_Topic_Attention_Polarity.png', bbox_inches='tight') #, dpi=400

#ax.figure.savefig('Sentiment Weighted Topic Attention Polarity.png', bbox_inches='tight')
```
